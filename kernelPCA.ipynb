```python
import pandas as pd
import matplotlib.pyplot as plt
from decomposition import CustomKernelPCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
```

```python
spiral_data = pd.read_csv("dataset/TwoSpirals.csv", header=None, delimiter="\t").to_numpy()
spiral_data[:, -1] = [i if i > 0 else 0 for i in spiral_data[:, -1]]
spiral_data
```

```python
plt.scatter(spiral_data[:, 0], spiral_data[:, 1], c=spiral_data[:, 2])
```

```python
log_reg = LogisticRegression()
log_reg.fit(spiral_data[:, :-1], spiral_data[:, -1])
```

```python
y_pred = log_reg.predict(spiral_data[:, :-1])
print(f"F1-score - {f1_score(spiral_data[:, -1], y_pred)}")
```

```python
kernal_pca = CustomKernelPCA()
data = kernal_pca.fit_transform(spiral_data[:, :-1])
```

```python
for num in range(5, len(spiral_data), 5):
    transform_data = data[:, :num]
    log_reg.fit(transform_data, spiral_data[:, -1])
    y_pred = log_reg.predict(transform_data)
    print(f"Dimension - {num}, F1 score - {f1_score(spiral_data[:, -1], y_pred)}")
```
